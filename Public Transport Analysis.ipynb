# PUBLIC TRANSPORT ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42) # Buat data set

tahun = list(range(2020, 2026))
bulan = list(range(1, 13))
moda = ['MRT', 'LRT', 'Transjakarta', 'KRL', 'Bus']

data = []
for t in tahun:
    for m in moda:
        for b in bulan:
            pengguna = np.random.randint(30000, 300000)
            data.append([t, m, b, pengguna])
            
df = pd.DataFrame(data, columns=['Tahun', 'Bulan', 'Moda', 'Jumlah Pengguna'])
#df = df.head(200)
df.head()
```

### Visualisasi Tren Tahunan

```python
df_grouped = df.groupby(['Tahun', 'Moda'])['Jumlah Pengguna'].sum().reset_index()
df_grouped.head()
```

```python
df_grouped['Tahun'] = df_grouped['Tahun'].astype(str)

plt.figure(figsize=(10,6))
sns.lineplot(data=df_grouped, x='Tahun', y='Jumlah Pengguna', hue='Moda', marker='o')
plt.title('Jumlah Pengguna Transportasi Umum Tahun 2020-2024')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pengguna')
plt.grid(True)
plt.tight_layout()
plt.legend(title='Moda Transportasi', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

### 🧠 Insight:
- Terlihat adanya tren peningkatan jumlah pengguna dari tahun ke tahun.
- Moda TransJakarta dan KRL mendominasi penggunaan.
- Peningkatan ini menunjukkan pemulihan mobilitas masyarakat pasca pandemi COVID-19.
- Data ini bisa dijadikan indikator pertumbuhan ekonomi di kota X.
